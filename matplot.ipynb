```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#line plot
x=[1,2,3,4,5]
y=[10,49,67,75,90]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()
```

```python
import pandas as pd 
```

```python
#bar plot
data=pd.DataFrame({
     "class":['A','B','C','D','E'],
     "student":[57,45,33,55,12]
})
sns.barplot(x='class', y='student',data=data)
plt.title("Bar plot")
plt.show()
```

```python
#count plot
sns.countplot(x='class',data=data)
plt.title("count plot")
plt.show()
```

```python
#histogram
import numpy as np
data=np.random.randn(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
# Or using seaborn 
sns.histplot(data, bins=10, kde=True) 
plt.title("Histogram with KDE") 
plt.show()
```

```python
sns.kdeplot(data, shade=True) 
plt.title("KDE Plot") 
plt.show()
```

```python
sns.boxplot(y=data) 
plt.title("Box Plot") 
plt.show() 
```

```python
sns.violinplot(y=data) 
plt.title("Violin Plot") 
plt.show() 
```

```python
x = [1, 2, 3, 4, 5] 
y = [5, 4, 7, 10, 8] 
plt.scatter(x, y) 
plt.title("Scatter Plot") 
plt.show()
```

```python
# Seaborn way 
sns.scatterplot(x=x, y=y) 
plt.title("Scatter Plot") 
plt.show() 
```

```python
df = sns.load_dataset('iris') 
sns.pairplot(df, hue='species') 
plt.show()
```

```python
corr = df.corr() 
sns.heatmap(corr, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap") 
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
#devided dataset as independent (input)and (output) variable
x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
x
```

```python
y
```

```python
#s
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_train
```

```python
#introduce slr model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#predict
result = reg.predict([[10]])
result
```

```python
result= reg.predict(x_test)
result
```

```python
y_test
```

```python
# visualize plot
plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='pink')
plt.show()
```

```python
#for user input
years = int(input("Enter years of exp.:"))
myinput = pd.DataFrame(columns=['YearsExperience'],data=[[years]])
result = reg.predict(myinput)
print("Your Salary must be close to",round(result[0,0]))
```
